```python
prop64name = "PROPOSITION 063- FIREARMS. AMMUNITION SALES. INTIATIVE STATUTE."
```

```python
prop64data = props[props.prop_name == prop64name]
```

```python
prop64data.head()
```

```python
prop64data.describe()
```

```python
prop64data.info()
```

```python
contribs.info()
```

```python
prop64_with_contribs = pd.merge(prop64data, contribs, on="calaccess_committee_id")
```

```python
prop64_with_contribs.head()
```

```python
prop64_with_contribs.info()
```

```python
prop64_with_contribs.amount.sum()
```

```python
prop64_with_contribs.committee_position.value_counts()
```

```python
support = prop64_with_contribs[prop64_with_contribs.committee_position == "SUPPORT"]
oppose = prop64_with_contribs[prop64_with_contribs.committee_position == 'OPPOSE']
```

```python
support.info()
```

```python
oppose.info()
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / prop64_with_contribs.amount.sum()
```

```python
oppose.amount.sum() / prop64_with_contribs.amount.sum()
```

```python
merged = prop64_with_contribs
```

```python
merged.sort_values(by="amount", ascending=False).head()
```

```python
support.sort_values(by="amount", ascending=False).head()
```

```python
oppose.sort_values(by="amount", ascending=False).head()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values(by="amount", ascending=False)
```

```python
merged.groupby(['contributor_firstname', 'contributor_lastname']).amount.sum().reset_index().sort_values(by="amount", ascending=False)
```

```python
merged.groupby(['contributor_firstname', 'contributor_lastname', 'committee_position']).amount.sum().reset_index().sort_values(by="amount", ascending=False).head(10)
```
